```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
fear = pd.read_csv("fear_greed_index.csv")
trader = pd.read_csv("historical_data.csv")
fear.head()
trader.head()
fear.shape
trader.shape
fear.columns
trader.columns
fear.info()
trader.info()
fear.isnull().sum()
trader.isnull().sum()
fear = fear.dropna()
trader = trader.dropna()
fear.duplicated().sum()
trader.duplicated().sum()
fear = fear.drop_duplicates()

trader = trader.drop_duplicates()
fear['date'] = pd.to_datetime(fear['date'])
trader['Timestamp'] = pd.to_datetime(trader['Timestamp'])
fear['Date'] = fear['date'].dt.date

trader['Date'] = trader['Timestamp'].dt.date
fear.head()
trader.head()
merged = pd.merge(
    trader,
    fear[['Date', 'classification']],
    on='Date',
    how='left'
)
merged.head()
merged.to_csv("merged_data.csv", index=False)
```
